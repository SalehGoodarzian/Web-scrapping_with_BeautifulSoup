# Movie Dataset Creation

### Now we first find the links for all movies in the Wikipedia page for Walt Disney Pictures Films. Then, apply the above task of extracting information for their corresponding Wikiperdia page. This will be repeating the above procedure for more 400 times as there are more than 400 movies in the Walt Disney Wikipedia webpage

### Import necessary libraries

```python
import requests
from bs4 import BeautifulSoup as bs
import re
from datetime import datetime
```

### Request the content to obtain links for all movies in the Walt Disney Wikipedia webpage
We find some tables which must be extracted first

```python
url = 'https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films'
r = requests.get(url)
soup = bs(r.content, 'html.parser')

dates = soup.find_all('h3')
date = [w.find('span', class_='mw-headline').text for w in dates if w.find('span', class_='mw-headline')]
tables = soup.find_all('table', class_='wikitable')
```

### We go thru each table and extract the link

```python
movies = []
for table in tables:
    trs = table.find_all('tr')[1:]
    for tr in trs:
        if tr.a:
            movies.append('https://en.m.wikipedia.org/{}'.format(tr.a['href']))
```

We show forst 20 links below

```python
for i in movies[:20]:
    print(i)
```

### Now, using the procedure explained in the BeautifulSoup_exercise code, we extract the info for each of the links in the movies (that is a list)

```python
for url in movies[:5]:
    print(url)
    r = requests.get(url)
    soup = bs(r.content, 'html.parser')
    for tag in soup.find_all(['sup', 'span']):
        tag.decompose()

    if soup.find('table', class_='infobox'):
        data1 = soup.find('table', class_='infobox')
        trs = data1.find_all('tr')[2:]
        for tr in trs:
            if tr.find('th'):
                task = tr.find('th').get_text(' ', strip=True)
            else:
                task = ''
            if tr.find('li'):
                set = tr.find_all('li')
                person = [w.text.replace('\xa0', ' ') for w in set]
            elif tr.find('td'):
                person = tr.find('td').get_text(' ', strip=True)
            else:
                person = ''
            print(task + ':', person)

            if 'running' in task.lower():
                person2 = int(person.split()[0])
                print(task + ':', person2)

            if 'budget' in task.lower() or 'box' in task.lower():
                items = person.split()
                a = items[0].replace(',', '')
                a = float(re.findall('\d+(?:\.\d{1,})?', a)[0])
                if len(items) >= 2:
                    if items[1].lower() == 'million':
                        a = a * 1000000
                print(task + ':', a)

            if 'date' in task.lower():
                pp = []
                for i in person:
                    aa = i.split('(')[0].strip()
                    print(aa)
                    aa = datetime.strptime(aa, '%B %d, %Y')
                    pp.append(aa)
                    person = pp
                print(task + ':', person)

    print('\n========================')
```
